```python
req = Request('https://www.yellowpagesdirectory.com/', headers={'User-Agent': 'Mozilla/5.0'})
html_page = urlopen(req).read()
soup = BeautifulSoup(html_page, 'html.parser')
states_links = []
for link in soup.find_all('a', attrs={'href': re.compile("^https://www.yellowpagesdirectory.com/states/")}):
    states_links.append(link.get('href'))
# print(states_links)
```

```python
states_links_csv = states_links
     
# dictionary of lists  
dict = {'states_links_csv': states_links_csv}  
       
df = pd.DataFrame(dict) 
    
# saving the dataframe 
df.to_csv('states_links_csv.csv') 
```

```python
cities_states_links_csv = cities_states_links
     
# dictionary of lists  
dict = {'cities_states_links_csv': cities_states_links_csv}  
       
df = pd.DataFrame(dict) 
    
# saving the dataframe 
df.to_csv('cities_states_links_csv.csv') 
```

```python
cities_states_links_csv_df = pd.read_csv('cities_states_links_csv.csv')
cities_states_links_csv_df = cities_states_links_csv_df['cities_states_links_csv']
cities_states_links_csv_dfs = cities_states_links_csv_df.values.tolist()
```

```python
business_cities_states_links = []
for cities_states_links_csv_df in cities_states_links_csv_dfs:
    print(cities_states_links_csv_df)
    time.sleep(2)
    req_states_cities = Request(cities_states_links_csv_df, headers={'User-Agent': 'Mozilla/5.0'})
    html_page_states_cities = urlopen(req_states_cities).read()
    soup_states_cities = BeautifulSoup(html_page_states_cities, 'html.parser')
    re_compile = "^"+cities_states_links_csv_df+'/'
    for link in soup_states_cities.find_all('a', attrs={'href': re.compile(re_compile)}):
        business_cities_states_links.append(link.get('href'))
#         print(business_cities_states_links)
```

```python
business_cities_states_links_csv_df = pd.read_csv('business_cities_states_links_csv.csv')
business_cities_states_links_csv_df = business_cities_states_links_csv_df['business_cities_states_links_csv']
business_cities_states_links_csv_dfs = business_cities_states_links_csv_df.values.tolist()
```

```python
specific_business_cities_states_links = []
for business_cities_states_links_csv_df in business_cities_states_links_csv_dfs:
    print(business_cities_states_links_csv_df)
    time.sleep(1)
    req_states_cities_business = Request(business_cities_states_links_csv_df, headers={'User-Agent': 'Mozilla/5.0'})
    html_page_states_cities_business = urlopen(req_states_cities_business).read()
    soup_states_cities_business = BeautifulSoup(html_page_states_cities_business, 'html.parser')
    
    for soup_states_cities_business_badges in soup_states_cities_business.body.findAll('div', attrs={'class': 'new-nav'}):
#         print(soup_states_cities_business_badges)
        for badge in [soup_states_cities_business_badges]:
            phones = badge.findAll('span',attrs={'class': 'new-bg'})
            if(len(phones)>0):
                for phone in phones:
                    for phone_link in phone.findAll('a'):
                        specific_business_cities_states_links.append(phone_link.get('href'))
            else:
                for phone_link in badge.findAll('a'):
                    if(phone_link.get('href').find('http')==-1):
                        specific_business_cities_states_links.append('https://www.yellowpagesdirectory.com/'+phone_link.get('href'))
```
