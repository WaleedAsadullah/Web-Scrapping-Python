```python
name_specific_business_cities_states_link = []
phone_specific_business_cities_states_link = []
city_specific_business_cities_states_link = []
service_specific_business_cities_states_link = []
# 5350
for business_cities_states_links_csv_df in business_cities_states_links_csv_dfs[5350:]:
    print(business_cities_states_links_csv_df)
    other_deatils = business_cities_states_links_csv_df.split('/')
    city_name = other_deatils[3]
    service_name = other_deatils[4].replace('-',' ').replace('+',' ')
#     print(city_name)
#     print(service_name)
    time.sleep(1)
#     if(business_cities_states_links_csv_df.find('.com//')>0)
    req_states_cities_business = Request(business_cities_states_links_csv_df, headers={'User-Agent': 'Mozilla/5.0'})
    html_page_states_cities_business = urlopen(req_states_cities_business).read()
    soup_states_cities_business = BeautifulSoup(html_page_states_cities_business, 'html.parser')
    
    for soup_states_cities_business_badges in soup_states_cities_business.body.findAll('div', attrs={'class': 'new-nav'}):
        for badge in [soup_states_cities_business_badges]:
        #     print("--------------")
            phones = badge.findAll('span',attrs={'class': 'new-bg'})
            if(len(phones)>0):
#                 print("if")
                for phone in phones:
   #                     print("==================")
        #             print(phone)
                    badge_str = str(phone)
                    phone_num = badge_str.split('Phone:')
#                     print(phone_num[-1][:15])
                    phone_specific_business_cities_states_link.append(phone_num[-1][:15])
                    city_specific_business_cities_states_link.append(city_name)
                    service_specific_business_cities_states_link.append(service_name)
                    for phone_link in phone.findAll('a'):
                        for phone_link_children in phone_link.children:
                            name_specific_business_cities_states_link.append(phone_link_children)
#                             print(phone_link_children)
        #             for phone_link in phone.findAll('a'):
        #                 print(phone_link.get('href'))
            else:
#                 print("else")
                badge_str = str(badge)
                phone = badge_str.split('~')
#                 print(phone[-1][:15])
        #         print(badge)
                phone_specific_business_cities_states_link.append(phone[-1][:15])
                city_specific_business_cities_states_link.append(city_name)
                service_specific_business_cities_states_link.append(service_name)
                for phone_link in badge.findAll('a'):
        #             print(phone_link.children)
                    phone_link_children = phone_link.children
#                     print(phone_link_children)
                    for phone_link_children in phone_link.children:
#                         print("7777777")
#                         print(phone_link_children)
                        name_specific_business_cities_states_link.append(phone_link_children)
                    break
        #             print(phone_link.get('href').find('http'))
#                     if(phone_link.get('href').find('http')==-1):
#                         print('https://www.yellowpagesdirectory.com/'+phone_link.get('href'))

```
